```python
# Import all relevant libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Specify URL for getting S&P 500 stocks and parse them with BeautifulSoup
URL = "https://stockanalysis.com/list/sp-500-stocks/"
page = requests.get(URL)
soup = BeautifulSoup(page.text, "html.parser")
```

```python
# Find the table of stocks
table = soup.find('table')

# Extract data from the table
data = []
for row in table.find_all('tr'):
    row_data = []
    for cell in row.find_all(['th', 'td']):
        row_data.append(cell.text.strip())
    data.append(row_data)
```

```python
# Specify headers and data separately
headers = data[0]
data = data[1:]

# Create a dataframe with the table and index set as No.
df = pd.DataFrame(data,columns=headers)
df = df.set_index("No.")
df
```

```python
df.to_csv('stock tickers S&P 500.csv')
```
